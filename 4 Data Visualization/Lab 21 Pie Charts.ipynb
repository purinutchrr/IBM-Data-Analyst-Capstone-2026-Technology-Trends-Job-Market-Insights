### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
top5_db = df['DatabaseWantToWorkWith'].str.split(';').explode().value_counts().head()
```

```python
plt.figure(figsize = (8, 8))
plt.pie(top5_db, labels = top5_db.index, autopct = '%1.2f%%', startangle = 90)
plt.title('Top 5 Databases Respondents Want to Work With')
plt.show()
```

1.2 The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
top5_dev = df['DevType'].value_counts().head()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(top5_dev, labels = top5_dev.index, autopct = '%1.2f%%', startangle = 135)
plt.title('Top 5 Developer Roles')
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
##Write your code here
top_opsys = df['OpSysProfessional use'].str.split(';').explode().value_counts().head()
```

```python
plt.figure(figsize = (8, 8))
plt.pie(top_opsys, labels = top_opsys.index, autopct = '%1.2f%%', startangle = 135)
plt.title('Top Operating Systems Used by Respondents for Professional use')
plt.show()
```
